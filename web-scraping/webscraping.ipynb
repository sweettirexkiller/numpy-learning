```python
def interstingNumber(number, type):
    url = 'http://numbersapi.com/'+str(number)+'/'+type+'?json'
    page = requests.get(url)
    data=json.loads(page.text)
    return data['text']
```

```python
interstingNumber(5, 'trivia')
```

```python
def interstingDate(day,month):
    url = 'http://numbersapi.com/'+str(day)+'/'+str(month)+'?json'
    page = requests.get(url)
    data=json.loads(page.text)
    return data['text']
```

```python
def multileCuriosity(arrayOfNumbers, type):
    arrayOfNumbers = [str(i) for i in arrayOfNumbers]
    url = "http://numbersapi.com/"+(",".join(arrayOfNumbers)).strip() + "/"+ type + '?json'
    page = requests.get(url)
    data = json.loads(page.text)
    return json_normalize(data).T
    
```

```python
multileCuriosity([1,2,3],'math')
```
