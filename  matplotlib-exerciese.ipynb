```python
theta = np.linspace(0,2*np.pi,1000)

plt.figure(figsize=(8,8))
x = np.sin(theta)
y = np.cos(theta)
plt.axis('equal')
plt.fill(x,y,'k', facecolor='#FF0000',edgecolor='#990000', alpha=0.5)
plt.plot([0],[0], 'md', color='k')
```

```python
import datetime
now = datetime.datetime.now()
theta = np.linspace(0,2*np.pi,1000)

plt.figure(figsize=(8,8))
x = np.sin(theta)
y = np.cos(theta)
plt.axis('equal')
plt.fill(x,y,'k', facecolor='#FF0000',edgecolor='#990000', alpha=0.5)
plt.plot([0],[0], 'md', color='k')
plt.text(1, 0,"3", horizontalalignment='left', verticalalignment='top',fontsize=20)   
plt.text(0, 1,"12", horizontalalignment='left', verticalalignment='bottom',fontsize=20)
plt.text(-1, 0,"9", horizontalalignment='right', verticalalignment='top',fontsize=20)   
plt.text(0, -1,"6", horizontalalignment='left', verticalalignment='top',fontsize=20)

hourInAngles = 90 -(now.hour%12)*30
minuteInAngles = 90 - now.minute*6


yMinutes = np.sin(np.deg2rad(minuteInAngles))
xMinutes = np.cos(np.deg2rad(minuteInAngles))

yHours = np.sin(np.deg2rad(hourInAngles))
xHours = np.cos(np.deg2rad(hourInAngles))


plt.scatter(xMinutes,yMinutes, color='blue')

plt.scatter(xHours,yHours, color='black')

    
plt.show()
```

```python
cmap = matplotlib.cm.get_cmap('coolwarm')
n = 30
x = np.round(np.linspace(0,10,n))
colors = np.round(np.linspace(0,1,n),2)

for i in range(n):
    plt.fill([i, i+1, i+1, i], [0,0,1,1], facecolor=cmap(colors[i]))
    

plt.show()
```

```python
flights=sns.load_dataset("flights")
meanPassangers = flights["passengers"].groupby(flights["year"]).agg(np.mean)
meanPassangers
```

```python
plt.title('Wykres sredniej liczby pasazerow od roku', color='c')

ax = plt.subplot()

plt.plot(meanPassangers, "rv-.",color='m', label='srednia roczna liczba pasazerow')
plt.plot(np.linspace(1949,1960, 5000), np.linspace(370,370, 5000),"--",color='b')
plt.plot(np.linspace(1949,1960, 5000), np.linspace(220,220, 5000),"--",color='b')
plt.plot(np.linspace(1952,1952, 5000), np.linspace(0,460, 5000),"--",color='b')
plt.plot(np.linspace(1956,1956, 5000), np.linspace(0,460, 5000),"--",color='b')


plt.xlabel('Rok', color='y')
plt.ylabel('Liczba pasazerow', color='y')



plt.legend()
```

```python
tips = sns.load_dataset('tips')
```

```python
smoker = tips.smoker.value_counts()
ax1 = plt.subplot()
ax1.pie(smoker, labels=['No', 'Yes'],startangle = 66, colors= ['r', 'b'],explode = (0, 0.3), autopct="%1.2f%%")
ax1.axis('equal')
plt.title('Palacze')
ax1.legend(title='Legenda', facecolor='y',edgecolor='b')
plt.show()
```

```python
smoker = tips.smoker.value_counts()

ax1 = plt.subplot()
rects = ax1.bar([1,2], smoker, color=['c', 'm'], yerr=(5,5))
plt.axhline(100,color='k',linestyle='--')
plt.title('Palacze')
ax1.legend(rects, ['Nie', 'Tak'])
plt.xticks([1,2], ['Nie', 'Tak'], rotation=27)
ax1.grid(True, linestyle='--')
plt.show()
```

```python
ax = sns.boxplot(tips['day'], tips['total_bill'],tips['sex'], palette="Set1", linewidth=6)
ax.legend(title='Plec', facecolor='y',edgecolor='b')
ax.set_facecolor('w')
ax.grid(True, linestyle='--')
plt.show()
```
