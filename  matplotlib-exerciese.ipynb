```python
import numpy as np
import pandas as pd
import matplotlib
matplotlib.__version__
import matplotlib.pyplot as plt
import seaborn as sns
sns.__version__
```

```python
%matplotlib inline
```

```python
theta = np.linspace(0,2*np.pi,1000)

plt.figure(figsize=(8,8))
x = np.sin(theta)
y = np.cos(theta)
plt.axis('equal')
plt.fill(x,y,'k', facecolor='#FF0000',edgecolor='#990000', alpha=0.5)
plt.plot([0],[0], 'md', color='k')
plt.text(0, 0,"(0,0)", horizontalalignment='left', verticalalignment='top')   
plt.show()
```
