```python
x = np.linspace(-10,10,50)
y = x**2
plt.plot(x,y)
plt.show()
```

```python
t = np.linspace(0,2**np.pi, endpoint=False)
x = 16*np.sin(t)**3
y = 13*np.cos(t) - 5*np.cos(2*t) - 2*np.cos(3*t) - np.cos(4*t)
plt.fill(x,y,fill=True,color='r')
plt.show()
```

```python
t = np.linspace(0,2**np.pi,100, endpoint=False)
x = 16*np.sin(t)**3
y = 13*np.cos(t) - 5*np.cos(2*t) - 2*np.cos(3*t) - np.cos(4*t)
plt.fill(x,y,fill=False,color='k')
plt.axis("equal")
plt.xticks([-5,-1.5,0,1.5,5],['a','b','0','b','a'])
plt.yticks([-17,0,5])

plt.show()
```

```python
theta = np.linspace(0,np.pi**2,100)
y = 3+2*np.sin(theta)
x = 3+2*np.cos(theta)
plt.plot(x,y,'k')
```

```python
p = [0.3,0.25,0.4,0.05]
e = ['no','yes','i told you','pay I will tell']
x = np.arange(len(p))
s = 0.75
plt.bar(x,p,s,color=['r','g','b','y'])
plt.xticks(x,e,rotation=20)
plt.show()
```

```python
p = [0.3,0.25,0.4,0.05]
e = ['no','yes','i told you','pay I will tell']
x = np.arange(len(p))
s = 0.75
plt.title('How are you today ?')
plt.pie(p,labels=e,colors=['r','g','b','y'],startangle=55)
plt.show()
```

```python
import matplotlib.mlab as mlab

mu,sigma = 100,15

x = mu + sigma*np.random.randn(10000)

n,bins,patches = plt.hist(x,20,density=1, facecolor='green', alpha=0.75,edgecolor='r')

plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title(r'$\mathrm{Histogram\ of\ IQ:} \mu=100, \sigma=15$')
plt.axis([40,160,0,0.03])
plt.grid(True)
plt.show()
```

```python
tips = sns.load_dataset('tips')
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.lmplot(x='sepal_length', y='sepal_width',data=iris, hue='species', fit_reg=False)
plt.axis('equal')
plt.show()
```

```python
sns.lmplot(x='sepal_length', y='sepal_width',data=iris, hue='species', fit_reg=True)
plt.axis('equal')
plt.show()
```

```python
sns.pairplot(iris, diag_kind='kde',hue='species',vars=['sepal_length', 'sepal_width'])
```

```python
flights = sns.load_dataset('flights')
flights.columns
```

```python
plt.figure(figsize=(6,6))
flights_pivot = flights.pivot('month', 'year','passengers')
sns.heatmap(flights_pivot, annot=True,fmt='d',linewidths=0.5)
plt.show()
```
